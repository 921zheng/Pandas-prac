### 4.use series create data

```python
brand=pd.Series(['A','B','C','D'])
color=pd.Series(['red','yellow','blue','brown'])
df=pd.DataFrame({'brand':brand, 'color':color})
df
```

### 4.1 change index

```python
df=pd.DataFrame({'brand':brand, 'color':color}, index=[4,6,2,1])
df
```
